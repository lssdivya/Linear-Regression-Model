```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
df.rename(columns={'per capita income (US$)': 'per_capita_income'}, inplace=True)
```

```python
df.head()
```

```python
plt.xlabel('year')
plt.ylabel('Per Capita Income (US$)')
plt.scatter(df.year, df.per_capita_income, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df.per_capita_income)
```

```python
year_to_predict = pd.DataFrame({"year": [2020]})
prediction = reg.predict(year_to_predict)
print(f"Predicted per-capita income for 2020: ${prediction[0]:,.2f}")
```

```python
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.scatter(
    df["year"], df["per_capita_income"], color="red", marker="+", label="Actual data")
plt.plot(
    df["year"], reg.predict(df[["year"]]), label="Regression line")
plt.legend()
plt.show()
```

This linear regression model predicts that Canada's per-capita income in 2020 will be approximately $41,288.69. his prediction is based on the historical relationship between year and per-capita income in the dataset. Since linear regression assumes a straight-line trend, the actual value may differ because economic conditions can change over time.
